## THỰC HÀNH 2.8: 
---
**Yêu cầu 1:** Sinh viên đọc ảnh images/Thuchanh2_8.jpeg, dạng xám và hiển thị ảnh

**Yêu cầu 2:** Sử dụng phương pháp làm mịn trung bình với kernel 5x5, 10x10, 20x30 và hiển thị kết quả

**Yêu cầu 3:** Sử dụng phương pháp làm mịn trung vị với k =  7, 11, 21 và hiển thị kết quả

**Yêu cầu 4:** Sử dụng phương pháp làm mịn Gaussian với kernel = (7,7) và sigmaY, sigmaY lần lượt là (0,0); (0,20); (20,0) và hiển thị kết quả

<img src='images/Thuchanh2_8.jpeg' width='400px'>

```python
#{SINH VIÊN VIẾT CODE VÀO ĐÂY}

img_28 = cv2.imread('images/Thuchanh2_8.jpeg',0)

plt.imshow(img_28,cmap='gray')
plt.show()

```

```python
#Làm mịn trung bình
#Làm mịn ảnh với các kernel khác nhau:
#kernel = 5 x 5:
img_1 = cv2.blur(img_28,(5,5))

#kernel = 10 x 10:
img_2 = cv2.blur(img_28,(10,10))

#kernel = 20 x 30:
img_3 = cv2.blur(img_28,(20,30))


#Hiển thị ảnh gốc và ảnh làm mờ với các kernel khác nhau:
plt.figure(figsize=(20,30))
plt.subplot(2, 4, 1)
plt.imshow(img_28,cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(2, 4, 2)
plt.imshow(img_1,cmap='gray')
plt.title('kernel = 5 x 5')

plt.subplot(2, 4, 3)
plt.imshow(img_2,cmap='gray')
plt.title('kernel = 10 x 10')

plt.subplot(2, 4, 4)
plt.imshow(img_3,cmap='gray')
plt.title('kernel = 20 x 30')


plt.show()
```

```python
#Làm mịn ảnh với các kernel khác nhau sử dụng bộ lọc trung vị:
#ksize = 7:
median_1 = cv2.medianBlur(img_28,7)

#ksize = 11:
median_2 = cv2.medianBlur(img_28,11)

#ksize =  21:
median_3 = cv2.medianBlur(img_28,21)


#Hiển thị kết quả:
plt.figure(figsize=(20,30))
plt.subplot(2, 4, 1)
plt.imshow(img_28,cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(2, 4, 2)
plt.imshow(median_1,cmap='gray')
plt.title('Median = 7')

plt.subplot(2, 4, 3)
plt.imshow(median_2,cmap='gray')
plt.title('Median = 11')

plt.subplot(2, 4, 4)
plt.imshow(median_3,cmap='gray')
plt.title('Median = 21')

plt.show()
```

**Yêu cầu 2:** Sinh viên đọc ảnh images/Thuchanh2_9_2.png và hiển thị ảnh. Lựa chọn phương pháp khử nhiễu và các tham số phù hợp để thu được ảnh khử nhiễu tốt nhất. Lưu ảnh đã xử lý vào thư mục images/Saves/MSV_plane.png

```python
#{SINH VIÊN VIẾT CODE VÀO ĐÂY}

img_29_1 = cv2.imread('images/Thuchanh2_9_1.jpeg')

ksize = 5

#Làm mờ với phương thức MedianBlur:
img_median = cv2.medianBlur(img_29_1,ksize) #ksize: chỉ có một số và phải là số lẻ > 1

#Hiển thị kết quả
plt.figure(figsize=(10,8))
plt.subplot(1, 2, 1)
plt.imshow(img_29_1, cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1, 2, 2)
plt.imshow(img_median, cmap='gray')
plt.title('Ảnh làm mờ MedianBlur()')

plt.show()
```

```python
#Lưu ảnh 
cv2.imwrite('images/Saves/1921050540_brain.png',img_median)
```

```python
#Thiết lập mặt nạ lọc:
img_29_2 = cv2.imread('images/Thuchanh2_9_2.png')

kernel = (15,15)

#Làm mịn với phương thức Gaussianblur():
img_gauss = cv2.GaussianBlur(img_29_2,kernel,0)

#Hiển thị kết quả:
plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
plt.imshow(img_29_2, cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(2, 2, 2)
plt.imshow(img_gauss, cmap='gray')
plt.title('Ảnh lọc Gaussian')
plt.show()
```

```python
#Lưu ảnh 
cv2.imwrite('images/Saves/1921050540_plane.png',img_gauss)
```

## THỰC HÀNH 2.10: 
---
Yêu cầu :
* Sinh viên đọc và hiển thị ảnh images/Thuchanh2_10. ở chế độ mặc định.
* Áp dụng 4 bộ lọc khác nhau trên ảnh và hiển thị kết quả (xem chi tiết các bộ lọc trong slide bài giảng)
<img src = 'images/Thuchanh2_10.png' width='300px'>

https://setosa.io/ev/image-kernels/

```python
#{SINH VIÊN VIẾT CODE VÀO ĐÂY}

img_210 = cv2.imread('images/Thuchanh2_10.png')


plt.imshow(img_210)
plt.show()
```

```python
#1. Bộ lọc blur:
#Khởi tạo kernel:
kernel_1 = np.array([[0.0625,0.125,0.0625], 
                            [0.125,0.25,0.125],
                            [0.0625,0.125,0.0625]])
kernel_1


#Thực hiện lọc ảnh với kernel ở trên:
img_210_1 = cv2.filter2D(img_210, -1, kernel_1)

#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(img_210,cmap='gray')
plt.title('Ảnh gốc ban đầu')


plt.subplot(1,2,2)
plt.imshow(img_210_1,cmap='gray')
plt.title('Ảnh qua bộ lọc blur')
plt.show()
```

```python
#2. Bộ lọc emboss
#Khởi tạo kernel:
kernel_2 = np.array([[-2,-1,0], 
                        [-1, 1,1],
                        [0,1,2]])
kernel_2

#Thực hiện lọc ảnh với kernel ở trên:
img_210_2 = cv2.filter2D(img_210, -1, kernel_2)

#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(img_210,cmap='gray')
plt.title('Ảnh gốc ban đầu')


plt.subplot(1,2,2)
plt.imshow(img_210_2,cmap='gray')
plt.title('Ảnh qua bộ lọc emboss')
plt.show()
```

```python
#3. Bộ lọc bottom soble:
#Khởi tạo kernel:
kernel_3 = np.array([[ -1,-2, -1], 
                           [0, 0,0],
                           [ 1,2,1]])
kernel_3


#Thực hiện lọc ảnh với kernel ở trên:
img_210_3 = cv2.filter2D(img_210, -1, kernel_3)

#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(img_210,cmap='gray')
plt.title('Ảnh gốc ban đầu')


plt.subplot(1,2,2)
plt.imshow(img_210_3,cmap='gray')
plt.title('Ảnh qua bộ lọc bottom soble')
plt.show()
```

```python
#4. Bộ lọc outline):
#Khởi tạo kernel:
kernel_4 = np.array([[-1,-1, -1], 
                        [-1, 8,-1],
                        [-1, -1,-1]])

kernel_4

#Thực hiện lọc ảnh với kernel ở trên:
img_210_4 = cv2.filter2D(img_210, -1, kernel_4) 

#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(img_210,cmap='gray')
plt.title('Ảnh gốc ban đầu')


plt.subplot(1,2,2)
plt.imshow(img_210_4,cmap='gray')
plt.title('Ảnh qua bộ outline')
plt.show()
```
