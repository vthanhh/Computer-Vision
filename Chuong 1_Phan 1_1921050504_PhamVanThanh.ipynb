```python
import os

def renamefile(path,name_new):
    for count, filename in enumerate(os.listdir(path)):
        src = path + filename
        dst = path + name_new + str(count) + ".jpg"
        os.rename(src, dst)
```

```python
path_tulips = 'images/tulips/'
name = 'Tulips_'

renamefile(path_tulips,name)
```

```python
import pathlib

pic_tulips = list(pathlib.Path(path_tulips).glob('*'))
print('Tổng số ảnh trong thư mục: ',len(pic_tulips))
print(pic_tulips[0])
```

```python
plt.figure(figsize = (16,18))
for count, filename in enumerate(pic_tulips):
    img = cv2.imread(str(filename))
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    ax = plt.subplot(5,4,count+1)
    
    plt.imshow(img)
    
    plt.title(str(filename).split('\\')[-1])
    plt.grid()
    plt.axis('on')

plt.show()
```
