## THỰC HÀNH 2.6: 
---
**Yêu cầu 1:** Sinh viên đọc ảnh images/Thuchanh2_6.jpg ở dạng xám

**Yêu cầu 2:** Thực hiện cắt ngưỡng ảnh với (Type = cv2.THRESH_BINARY) lần lượt với T = 50, 100, 150. Hiển thị kết quả như slide minh hoạ. Lưu ảnh vào thư mục Saves

**Yêu cầu 3:** Thiết lập T = 75. Thực hiện cắt ngưỡng với Type = cv2.THRESH_TRUNC, cv2.THRESH_BINARY_INV, cv2.THRESH_TOZERO. Hiển thị kết quả như slide minh hoạ. Lưu ảnh vào thư mục Saves
<img src='images/Thuchanh2_6.jpg' width='300px'>

```python
#{SINH VIÊN VIẾT CODE Ở ĐÂY}

img_26 = cv2.imread('images/Thuchanh2_6.jpg',0)

ret,pic1 = cv2.threshold(img_26,50,255,cv2.THRESH_BINARY)
ret,pic2 = cv2.threshold(img_26,100,255,cv2.THRESH_BINARY)
ret,pic3 = cv2.threshold(img_26,150,255,cv2.THRESH_BINARY)

#Hiển thị kết quả:
plt.figure(figsize=(20,15))
plt.subplot(2, 4, 1)
plt.imshow(img_26, cmap='gray')
plt.title('Ảnh gốc ban đầu')

plt.subplot(2, 4, 2)
plt.imshow(pic1, cmap='gray')
plt.title('Ảnh cắt ngưỡng T = 50')

plt.subplot(2, 4, 3)
plt.imshow(pic2, cmap='gray')
plt.title('Ảnh cắt ngưỡng T = 100')

plt.subplot(2, 4, 4)
plt.imshow(pic3, cmap='gray')
plt.title('Ảnh cắt ngưỡng T = 150')

plt.show()
```

```python
#Lưu ảnh :
cv2.imwrite('images/Saves/1921050540_anhgoc.jpg',img_26)
cv2.imwrite('images/Saves/1921050540_t50.jpg',pic1)
cv2.imwrite('images/Saves/1921050540_t100.jpg',pic2)
cv2.imwrite('images/Saves/1921050540_t150.jpg',pic3)
```

```python
T = 75

ret,pic4 = cv2.threshold(img_26,T,255,cv2.THRESH_TRUNC)
ret,pic5 = cv2.threshold(img_26,T,255,cv2.THRESH_BINARY_INV)
ret,pic6 = cv2.threshold(img_26,T,255,cv2.THRESH_TOZERO)
#Hiển thị ảnh:
plt.figure(figsize=(18,8))
plt.subplot(1,4,1)
plt.imshow(img_26, cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1,4,2)
plt.imshow(pic4, cmap='gray')
plt.title('Ảnh cắt ngưỡng 75 THRESH_TRUNC')

plt.subplot(1,4,3)
plt.imshow(pic5, cmap='gray')
plt.title('Ảnh cắt ngưỡng 75 THRESH_BINARY_INV')

plt.subplot(1,4,4)
plt.imshow(pic6, cmap='gray')
plt.title('Ảnh cắt ngưỡng 75 THRESH_TOZERO')

plt.show()
```

```python
#Lưu ảnh:
cv2.imwrite('images/Saves/1921050540_t75_1.jpg',pic4)
cv2.imwrite('images/Saves/1921050540_t75_2.jpg',pic5)
cv2.imwrite('images/Saves/1921050540_t75_3.jpg',pic6)
```

## THỰC HÀNH 2.7: 
---
Yêu cầu :
* Yêu cầu 1: Sinh viên đọc ảnh images/Thuchanh2_6.jpg ở dạng ảnh xám. Thực hiện cân bằng sáng trên ảnh Xám, hiển thị ảnh gốc và ảnh sau cần bằng cùng với Histogram tương ứng của mỗi ảnh.
* Yêu cầu 2: Sinh viên đọc ảnh images/Thuchanh2_6.jpg ở dạng ảnh màu. Thực hiện cân bằng sáng trên ảnh màu, hiển thị ảnh gốc và ảnh sau cần bằng ở dạng ảnh màu.
* Yêu cầu 3: Lưu ảnh xám và ảnh màu đã cân bằng vào thư mục images/Saves.

<img src = 'images/Thuchanh2_6.jpg' width='150px'>

```python
#{SINH VIÊN VIẾT CODE Ở ĐÂY}

img_27 = cv2.imread('images/Thuchanh2_6.jpg',0)

#Thực hiện cân bằng sáng với phương thức equalizeHist():
img_equ_27 = cv2.equalizeHist(img_27)

#Hiển thị ảnh và Histogram tương ứng.
#Ảnh 1
plt.figure(figsize=(10,7))
plt.subplot(2, 2, 1)
plt.imshow(img_27, cmap='gray')
plt.title('Ảnh Gốc')

plt.subplot(2, 2, 3)
plt.hist(img_27.ravel(),256,[0,256])
plt.title('Histgram của ảnh gốc')

#Ảnh 2:
plt.subplot(2, 2, 2)
plt.imshow(img_equ_27, cmap='gray')
plt.title('Ảnh đã cân bằng sáng ')

plt.subplot(2, 2, 4)
plt.hist(img_equ_27.ravel(),256,[0,256])
plt.title('Histogram của ảnh đã cân bằng')
plt.show()
```

```python
img_27_2 = cv2.imread('images/Thuchanh2_6.jpg')

# đổi từ hệ màu BGR sang YUV:
img27_yuv = cv2.cvtColor(img_27_2,cv2.COLOR_BGR2YUV)
#Cân bằng sáng trên kênh Y:
img27_yuv[:,:,0] = cv2.equalizeHist(img27_yuv[:,:,0])

# đổi từ hệ YUV  về BGR:
img27_bgr = cv2.cvtColor(img27_yuv,cv2.COLOR_YUV2BGR)


plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_27_2,cv2.COLOR_BGR2RGB))
plt.title('Ảnh màu Gốc')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img27_bgr,cv2.COLOR_BGR2RGB))
plt.title('Ảnh màu đã cân bằng')

plt.show()
```

```python
#Lưu ảnh:
cv2.imwrite('images/Saves/1921050540_img_equ.jpg',img_equ_27)
cv2.imwrite('images/Saves/1921050540_img_bgr.jpg',img27_bgr)
```
