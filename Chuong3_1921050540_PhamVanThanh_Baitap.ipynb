## THỰC HÀNH 3.1
---

1. Đọc và hiển thị ảnh Thuchanh3_1.jpg ở dạng ảnh xám
2. Sử dụng bộ lọc Sobel theo hướng x, y và XY. hiển thị kết quả.
3. Sử dụng phương thức Sobel với ksize = 7 theo X, Y và XY. Hiển thị kết quả.

<img src='images/Thuchanh3_1.jpg' width='500px'>

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

#Đọc ảnh ở chế độ ảnh xám:
img_31 = cv2.imread('images/Thuchanh3_1.jpg',0)

#Hiển thị ảnh gốc:
plt.imshow(img_31,cmap='gray')
plt.show()
```

```python
#Khởi tạo kernel Sobel theo hướng X:
sobel_x = np.array([[-1, 0, 1], 
                    [-2, 0, 2],
                    [-1, 0, 1]])
sobel_x
```

```python
#Thực hiện lọc ảnh với kernel ở trên:
img_x = cv2.filter2D(img_31, -1, sobel_x)

#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(img_31,cmap='gray')
plt.title('Ảnh gốc ban đầu')

plt.subplot(1,2,2)
plt.imshow(img_x,cmap='gray')
plt.title('Sobel_X')
plt.show()
```

```python
#Khởi tạo kernel sobel theo hướng Y:
sobel_y = np.array([[-1,-2,-1], 
                    [ 0, 0, 0],
                    [ 1, 2, 1]])
sobel_y
```

```python
#Thực hiện lọc ảnh với kernel ở trên:
img_y = cv2.filter2D(img_31, -1, sobel_y)

#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(img_31,cmap='gray')
plt.title('Ảnh gốc ban đầu')


plt.subplot(1,2,2)
plt.imshow(img_y,cmap='gray')
plt.title('Sobel_Y')
plt.show()
```

```python
#kết hợp theo X, Y
img_xy = img_x + img_y

#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_x,cmap='gray')
plt.title('Sobel_X')

plt.subplot(1,3,2)
plt.imshow(img_y,cmap='gray')
plt.title('Sobel_Y')

plt.subplot(1,3,3)
plt.imshow(img_xy,cmap='gray')
plt.title('Sobel_XY')
plt.show()
```

```python
#Sobel edges:
sobel_x = cv2.Sobel(img_31,cv2.CV_64F,1,0,ksize=7)
sobel_y = cv2.Sobel(img_31,cv2.CV_64F,0,1,ksize=7)

#Sobel theo cả x, y
img_sobel = sobel_x + sobel_y

#Hiển thị ảnh gốc và ảnh thay đổi kích thước
plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
plt.imshow(sobel_x,cmap='gray')
plt.title('Sobel theo X')

plt.subplot(1, 3, 2)
plt.imshow(sobel_y,cmap='gray')
plt.title('Sobel theo Y')

plt.subplot(1, 3, 3)
plt.imshow(img_sobel,cmap='gray')
plt.title('Sobel theo XY')
plt.show()
```

```python
sobel_xy = cv2.Sobel(img_31,-1,1,1,ksize=7)

plt.imshow(sobel_xy, cmap='gray')
plt.show()
```

## THỰC HÀNH 3.2
---
1. Đọc ảnh và hiển thị ảnh Thuchanh3_2.jpeg ở dạng ảnh xám
2. Sử dụng bộ lọc Prewitt và Roberts phát hiện cạnh, hiển thị kết quả.
3. Sử dụng bộ lọc Robinson (trong Slide) phát hiện cạnh và hiển thị kết quả.

<img src='images/Thuchanh3_2.jpeg' width='300px'>

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
#Đọc ảnh:
img_32 = cv2.imread('images/Thuchanh3_2.jpeg',0)

#Hiển thị ảnh gốc:
plt.imshow(img_32,cmap='gray')
plt.show()
```

```python
#Khởi tạo kernel prewitt theo hướng x:
prewitt_x = np.array([[-1, 0, 1], 
                     [ -1, 0, 1],
                     [ -1, 0, 1]])

#Khởi tạo kernel prewitt theo hướng y:
prewitt_y = np.array([[-1, -1,-1], 
                     [  0,  0, 0],
                     [  1,  1, 1]])

#Thực hiện lọc ảnh với kernel ở trên:
img_p32_x = cv2.filter2D(img_32, -1, prewitt_x)

#Thực hiện lọc ảnh với kernel ở trên:
img_p32_y = cv2.filter2D(img_32, -1, prewitt_y)

#kết hợp prewitt theo X, Y
img_pxy = img_p32_x + img_p32_y
```

```python
#Khởi tạo kernel Roberts theo hướng X,Y:
roberts_x = np.array( [[1, 0 ],
                       [0,-1 ]] )
  
roberts_y = np.array( [[ 0, 1 ],
                       [ -1, 0 ]])

#Thực hiện lọc ảnh với kernel ở trên:
img_r32_x = cv2.filter2D(img_32, -1, roberts_x)

#Thực hiện lọc ảnh với kernel ở trên:
img_r32_y = cv2.filter2D(img_32, -1, roberts_y)

#kết hợp prewitt theo X, Y
img_rxy = img_r32_x + img_r32_y
```

```python
#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_32,cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1,3,2)
plt.imshow(img_pxy,cmap='gray')
plt.title('Prewitt edge detection')

plt.subplot(1,3,3)
plt.imshow(img_rxy,cmap='gray')
plt.title('Roberts Edge Detection')
plt.show()
```

```python
#Khởi tạo kernel robinson theo hướng x:
robinson_x = np.array([[-2, -1, 0], 
                     [ -1, 0, 1],
                     [ 0, 1, 2]])

#Khởi tạo kernel robinson theo hướng y:
robinson_y = np.array([[-1, 0,1], 
                     [  -2,  0, 2],
                     [  -1,  0, 1]])
img_rox = cv2.filter2D(img_32, -1, robinson_x)
img_roy = cv2.filter2D(img_32, -1, robinson_y)

img_roxy = img_rox + img_roy
#Hiển thị ảnh đã xử lý qua bộ lọc:
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_rox,cmap='gray')
plt.title('robinson_X')

plt.subplot(1,3,2)
plt.imshow(img_roy,cmap='gray')
plt.title('robinson_Y')

plt.subplot(1,3,3)
plt.imshow(img_roxy,cmap='gray')
plt.title('Roberts_XY')
plt.show()
```

```python
#Khởi tạo kernel robinson theo hướng a:
robinson_a = np.array([[0, 1, 2], 
                     [ -1, 0, 1],
                     [ -2, -1, 0]])

#Khởi tạo kernel robinson theo hướng b:
robinson_b = np.array([[1, 2,1], 
                     [  0,  0, 0],
                     [  -1,  -2, -1]])
#Khởi tạo kernel robinson theo hướng c:
robinson_c = np.array([[2, 1,0], 
                     [  1,  0, -1],
                     [  0,  -1, -2]])
#Khởi tạo kernel robinson theo hướng d:
robinson_d = np.array([[1, 0,-1], 
                     [  2,  0, -2],
                     [  1,  0, -1]])
#Khởi tạo kernel robinson theo hướng e:
robinson_e = np.array([[0, -2,-2], 
                     [  1,  0, -1],
                     [  2,  1, 0]])
#Khởi tạo kernel robinson theo hướng f:
robinson_f = np.array([[-1, -2,-1], 
                     [  0,  0, 0],
                     [  1,  2, 1]])


img_roa = cv2.filter2D(img_32, -1, robinson_a)
img_rob = cv2.filter2D(img_32, -1, robinson_b)
img_roc = cv2.filter2D(img_32, -1, robinson_c)
img_rod = cv2.filter2D(img_32, -1, robinson_d)
img_roe = cv2.filter2D(img_32, -1, robinson_e)
img_rof = cv2.filter2D(img_32, -1, robinson_f)



#Hiển thị ảnh đã xử lý qua bộ lọc:

plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.imshow(img_roa,cmap='gray')
plt.title('robinson_a')

plt.subplot(2,3,2)
plt.imshow(img_rob,cmap='gray')
plt.title('roberts_b')

plt.subplot(2,3,3)
plt.imshow(img_roc,cmap='gray')
plt.title('robinson_c')

plt.subplot(2,3,4)
plt.imshow(img_rod,cmap='gray')
plt.title('robinson_d')

plt.subplot(2,3,5)
plt.imshow(img_roe,cmap='gray')
plt.title('robinson_e')

plt.subplot(2,3,6)
plt.imshow(img_rof,cmap='gray')
plt.title('robinson_f')
plt.show()
```

## THỰC HÀNH 3.3
---
1. Đọc và hiển thị ảnh Thuchanh3_3.png. 
2. Sử dụng phương pháp làm trơn ảnh medianBlur() để khử nhiễu.
3. Sử dụng bộ lọc Lapacian với mặt nạ số 2 trên ảnh đã làm trơn với ddepth = -1|CV_32FC4 và hiển thị kết quả

<img src = 'images/Thuchanh3_3.png' width = '300px'>

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}

img33 = cv2.imread('images/Thuchanh3_3.png',0)
plt.imshow(img33,cmap='gray')
plt.show()
```

```python
#Yêu cầu 1: Phát hiện cạnh biên của ảnh gốc chưa khử nhiễu
img_edge = cv2.Laplacian(img33,-1, 5)
plt.imshow(img_edge,cmap='gray')
plt.show()
```

```python
#Khử nhiễu ảnh:
img_new = cv2.medianBlur(img33, 5)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img33,cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1,2,2)
plt.imshow(img_new,cmap='gray')
plt.title('Ảnh sau khi khử nhiễu')
plt.show()
```

```python
img_33_1 = cv2.Laplacian(img_new,cv2.CV_64F,3)
img_33_2 = cv2.Laplacian(img_new,-1,7)

#Hiển thị kết quả:
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_new,cmap='gray')
plt.title('Ảnh gốc ban đầu')

plt.subplot(1,3,2)
plt.imshow(img_33_1,cmap='gray')
plt.title('Laplace ddepth=CV2_64F')

plt.subplot(1,3,3)
plt.imshow(img_33_2,cmap='gray')
plt.title('Laplace ddepth=-1')
plt.show()
```

# THỰC HÀNH 3.4
---
1. Đọc và hiển thị ảnh Thuchanh3_4.jpeg.
2. Sử dụng phương pháp Canny với các tham số phù hợp để thu được kết quả biên như trong slide.

<img src='images/Thuchanh3_4.jpeg' width='400px'>


```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
img_34 = cv2.imread('images/Thuchanh3_4.jpeg')

plt.imshow(img_34)
plt.show()
```

```python
#Tự động xác định ngưỡng - Auto canny:
sigma = 1.75
median = np.median(img_34)
lower = int(max(0,(1.0-sigma)*median))
upper = int(min(255,(1.0+sigma)*median))
auto_canny_34 = cv2.Canny(img_34,lower,upper,9)

#Hiển thị kết quả:
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)

plt.imshow(img_34,cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1, 2, 2)
plt.imshow(auto_canny_34,cmap='gray')
plt.title('Ảnh cạnh, biên với Canny')

plt.show()
```

```python
#làm mịn với Gaussian:
img_2 = cv2.GaussianBlur(img_34,(7,7),10)
img_edge = cv2.Canny(img_2, 20,45)


#Hiển thị kết quả:
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)

plt.imshow(img_34,cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1, 2, 2)
plt.imshow(img_edge,cmap='gray')
plt.title('Ảnh tách biên với Canny')
plt.show()
```
