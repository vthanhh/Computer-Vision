```python
#{SINH VIÊN VIẾT CODE Ở ĐÂY:}
#1. Đọc ảnh và hiển thị ảnh:
path_pic = 'images/Thuchanh_2_1.jpg'
img_rose = cv2.imread(path_pic)

#Chuyển sang hệ màu RGB để hiện thị:
img_rose = cv2.cvtColor(img_rose,cv2.COLOR_BGR2RGB)

plt.imshow(img_rose)
plt.show()
```

```python
#Chiều cao, chiều rộng và số pixel điểm ảnh:
h = img_rose.shape[0]
w = img_rose.shape[1]

print('Chiều cao và chiều rộng của ảnh :', h, 'x', w)
print('Số pixel điểm ảnh :', img_rose.size)
```

```python
#3. Thực hiện thay đổi kích thước theo yêu cầu sau:
#3.1. Thay đổi kích thước ảnh về 32x32, sử dụng tham số interpolation mặc định. 
# Hiển thị kết quả và Lưu lại ảnh vào thư mục Saves/MSV_img_31.jpg theo hệ màu BGR

dsize = (32,32) #Lưu ý thứ tự (Width, Height)

#Sử dụng phương thức resize():
img_rsize = cv2.resize(img_rose,
                         dsize, 
                         interpolation=cv2.INTER_LINEAR) 

#Hiển thị ảnh gốc và ảnh thay đổi kích thước
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)

plt.imshow(img_rose)
st = 'Ảnh gốc ban đầu:' + str(h) +  'x' + str(w)
plt.title(st)
plt.grid()

plt.subplot(1, 2, 2)
plt.imshow(img_rsize)
plt.title('Ảnh thay đổi kích thước :')
plt.grid()

plt.show()
```

```python
#Lưu ảnh sau resize:
cv2.imwrite('images/Saves/1921050540_img_31.jpg',cv2.cvtColor(img_rsize,cv2.COLOR_RGB2BGR))
```

```python
#3.2. Giữ nguyên chiều cao của ảnh gốc, giảm chiều rộng ảnh đi một nửa, 
#Sử dụng phương pháp INTER_NEAREST.
#Hiển thị kết quả và Lưu lại ảnh vào thư mục images/Saves/MSV_img_32.jpg theo hệ màu BGR

dsize2 = (490,735)
img_490x735 = cv2.resize(img_rose,dsize2,interpolation=cv2.INTER_NEAREST)

#Hiển thị ảnh gốc và ảnh thay đổi kích thước
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)

plt.imshow(img_rose)
st = 'Ảnh gốc ban đầu:' + str(h) +  'x' + str(w)
plt.title(st)
plt.grid()

plt.subplot(1, 2, 2)
plt.imshow(img_490x735)
plt.title('Ảnh thay đổi : ')
plt.grid()

plt.show()


#Lưu ảnh 
cv2.imwrite('images/Saves/1921050540_img_32.jpg',img_490x735)
```

```python
#3.3 Tăng chiều cao của ảnh lên gấp đôi, giữ nguyên chiều rộng của ảnh gốc; 
#Sử dụng phương pháp INTER_LANCZOS4. 
#Hiển thị kết quả và Lưu lại ảnh vào thư mục images/Saves/MSV_img_33.jpg theo hệ màu BGR

dsize3 = (980,1470)
img_980x1470 = cv2.resize(img_rose,dsize3,interpolation=cv2.INTER_LANCZOS4)

#Hiển thị ảnh gốc và ảnh thay đổi kích thước
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)

plt.imshow(img_rose)
st = 'Ảnh gốc ban đầu:' + str(h) +  'x' + str(w)
plt.title(st)
plt.grid()

plt.subplot(1, 2, 2)
plt.imshow(img_980x1470)
plt.title('Ảnh thay đổi : ')
plt.grid()

plt.show()


#Lưu ảnh 
cv2.imwrite('images/Saves/1921050540_img_33.jpg',img_980x1470)

```

## THỰC HÀNH 2.1 (Tiếp): 
---
**Yêu cầu 4:** Sinh viên sử dụng thư mục chứa các ảnh hoa hồng đã thực hiện trong bài thực hành 1.3 chương 1. Thực hiện thay đổi kích thước toàn bộ ảnh về độ phân giải 96x96 pixel. Lưu lại ảnh đã thay đổi vào thư mục mới có tên Roses_96 và đặt lại tên ảnh như sau: rose_96_i.jpg với i là các số thứ tự của ảnh.


```python
import os

def renamefile(path,name_new):
    for count, filename in enumerate(os.listdir(path)):
        src = path + filename
        dst = path + name_new + str(count) + ".jpg"
        os.rename(src, dst)
```

```python
path_roses = 'images/Roses/'
name = 'Rose_96_'

renamefile(path_roses,name)
```

```python
# {SINH VIÊN VIẾT CODE VÀO ĐÂY}
import pathlib
path_roses = 'images/Roses/'    #Thư mục chứa ảnh gốc
path_img_new = 'images/Roses_96/' #Thư mục chứa ảnh sau resize

#Lấy danh sách tên file có trong thư mục ảnh gốc vào một danh sách:
pic_roses = list(pathlib.Path(path_roses).glob('*'))

#Lấy số lượng ảnh trong thư mục:
num_pic = len(pic_roses)
```

```python
print('Tổng số ảnh trong thư mục ', path_roses, ' :',num_pic)
```

```python
#Thiết lập kích thước ảnh muốn resize:
d_h = int(input('Nhập vào chiều cao ảnh muốn thay đổi:'))
d_w = int(input('Nhập vào chiều rộng ảnh muốn thay đổi:'))
dsize = (d_w,d_h)
```

```python
plt.figure(figsize=(16,20))
hang = num_pic//5 + 1

for count, filename in enumerate(pic_roses):
    #Đọc từng ảnh theo chế độ mặc định
    img = cv2.imread(str(filename))
    
    #Thay đổi kích thước ảnh
    img_resize = cv2.resize(img,dsize, interpolation=cv2.INTER_AREA)
    
    #Ghi lại ảnh đã thay đổi vào thư mục mới, giữ nguyên tên ảnh cũ
    name =  str(filename).split("\\")[-1]   #Lấy tên ảnh gốc
    file_saves = path_img_new + name        #Tạo đường dẫn cho file ảnh mới
    
    #Lưu ảnh
    cv2.imwrite(file_saves,img_resize)
    
    #Hiển thị ảnh đã thay đổi:
    ax = plt.subplot(hang,5,count+1)
    plt.imshow(cv2.cvtColor(img_resize,cv2.COLOR_BGR2RGB))
    plt.title(name)
    plt.axis('on')
    
plt.show()
```

* Thực hiện lật ảnh theo chiều ngang và dọc

<img src = 'images/Thuchanh_2_3.jpg' width='300px'>

```python
#{SINH VIÊN VIẾT CODE VÀO ĐÂY}
#Đọc ảnh
path_pic = 'images/Thuchanh_2_3.jpg'
img = cv2.imread(path_pic)
img_23 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_23)
plt.show()
```

```python
#Xoay ảnh 90 độ theo chiều kim đồng hồ:
img_90_thuan = cv2.rotate(img_23,cv2.ROTATE_90_CLOCKWISE)
#Xoay ảnh 90 độ ngược chiều kim đồng hồ:
img_90_nguoc = cv2.rotate(img_23,cv2.ROTATE_90_COUNTERCLOCKWISE)

#Hiển thị ảnh đã xoay:
plt.imshow(img_90_thuan)
plt.show()
plt.imshow(img_90_nguoc)
plt.show()
```

```python
#Lật ảnh theo chiều dọc: flipCode = 0
img_lat_doc = cv2.flip(img_23, 0)
#Lật ảnh theo chiều ngang: flipCode > 0
img_lat_ngang = cv2.flip(img_23, 1)
#Hiển thị ảnh đã lật:
plt.imshow(img_lat_ngang)
plt.show()
plt.imshow(img_lat_doc)
plt.show()
```

```python
#Lưu ảnh vào thư mục images/Saves:

#Lưu ảnh vào thư mục images/Saves:
cv2.imwrite('images/Saves/xoaythuan.jpg',img_90_thuan)
cv2.imwrite('images/Saves/xoaynguoc.jpg',img_90_nguoc)
cv2.imwrite('images/Saves/latdoc.jpg',img_lat_doc)
cv2.imwrite('images/Saves/latngang.jpg',img_lat_ngang)
```
